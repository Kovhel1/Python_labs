# Задание 2. Среднее значение целых чисел

Написать программу (сценарий), сохраняющую сумму 1+2+3+4+5 в одну переменную, затем создающую другую переменную, содержащую среднее значение этих пяти чисел. Вывести фразу "Среднее значение этих 5 чисел равно" и результат работы с точностью до 5 знака после точки (т.е. до 1/100000).

```python
max_n = input("Введите максимальное число: ")
while(max_n.isdigit() == False):
  max_n = input("Это не число. Введите максимальное число: ")

max_n = float(max_n)
n = 0
sum = 0
max_n_int = round(max_n, 0)

while(max_n_int > n):
  n += 1
  sum += n
  
average5 = sum/n
print("среднее чисел от 1 до %.5f =  %.5f"  %(max_n_int, average5))


```

Добавьте в программу из предыдущего задания запрос на добавление ещё однонго числа к подсчёту среднего с помощью запуска интерактивной сессии функцией input(). После введения каждого нового числа пересчитывать и выводить резёльтата в формате: "Среднее значение этих (число чисел) чисел равно" и результат работы с точностью до 5 знака после точки (т.е. до 1/100000). Интерактивная сессия заканчивается после введения пользователем 0.

```python
count_n = 0
sum = 0

while(True):
 
  n = input("Введите число (0 = stop): ")
  while(n.isdigit() == False):
    n = input("Это не число. Введите число (0 = stop):  ")
  n = float(n)
  
  if n==0:
    break

  sum += n
  count_n += 1
  average5 = sum/count_n
  
  print("среднее =  %.5f"  %(average5))

```

Рассмотрим пример иллюстрирующий математическую модель описывающий траекторию кошачьего прыжка. Из второго закона Ньютона, предполагая отсутствие сопротивления воздуха, мы можем вывести зависимость вертикального (y) и горизонтального (x) положения котика от времени $t$:

$$x(t)=v_0cos(\alpha)t$$
$$y(t)=v_0sin(\alpha)t−\frac{1}{2}gt^2$$

где $v_0$ — начальная скорость, $g$ — ускорение свободного падения, значение которого положим равным 9.80665 $м/c^2$, $tg(\alpha) = y/x$.

Напишите программу которая сможет помочь котику понять следующие вещи:


1) как далеко и высоко он сможет прыгнуть при начальной скорости 5 м/с и направлении пражка $\alpha = 40^{\circ}$;

2) каким должен быть начальный угол для максимальной дальности прыжка;

3) каким должен быть начальный угол для максимальной высоты прыжка;

4) каким должны быть угол и скорость для попадания в точку с координатами (2,3) м через 1 с;

5) нарисуйте траекторию прыжка с помощью библиотеки matplotlib.


```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline

# Входные данные
def g():
   return 9.80665

print("Входные данные: \ng = ", g())
v0 = 10 #float(input("Начальная скорость v0: "))
a =  90 #float(input("Угол a(градусы): "))
a = math.radians(a)
max_t =  60 #float(input("Время прыжка t: "))

#1
def max_x(v0,a,max_t):
  return v0*math.cos(a)*max_t

def max_y(v0,a):
  return ((v0**2)*(math.sin(a)**2))/2*g() 

print("1. Длина и высота прыжка соответвенно:", max_x(v0,a,max_t),", ", max_y(v0,a))

#2
a_2_3 =  np.linspace(0.1, 90)

def max_ax(v0, a_2_3, max_t):
  a_2_3[0] = math.radians(a_2_3[0])
  max = max_x(v0,a_2_3[0],max_t)
  max_a = a_2_3[0]
  for a in a_2_3:
    a = math.radians(a)
    if(max_x(v0,a,max_t)>max):
      max = max_x(v0,a,max_t)
      max_a = a
  return math.degrees(a)

print("2. Угол для максимальной длины прыжка: ",  max_ax(v0, a_2_3, max_t))   


#3
def max_ay(v0, a_2_3):
  a_2_3[0] = math.radians(a_2_3[0])
  max = max_y(v0,a_2_3[0])
  max_a = a_2_3[0]
  for a in a_2_3:
    a = math.radians(a)
    if(max_y(v0,a)>max):
      max = max_y(v0,a)
      max_a = a
  return math.degrees(a)

print("3. Угол для максимальной  высоты прыжка: ",  max_ay(v0, a_2_3))  

#4 параметрически
def v0_4_1(a_2_3):
  v0 = []
  for a in a_2_3:
    a = math.radians(a)
    v0.append(2/math.cos(a)*1)
  return v0

v0_4_1 = v0_4_1(a_2_3)

def v0_4_2(a_2_3):
  v0 = []
  for a in a_2_3:
    a = math.radians(a)
    v0.append((3 + 0.5*g())/math.sin(a))
  return v0

v0_4_2 = v0_4_2(a_2_3)


plt.title("4.Решение системы уравнений")
plt.xlabel("v0_4_1") # ось абсцисс
plt.ylabel("v0_4_2") # ось ординат
plt.grid()    
plt.plot(a_2_3,v0_4_1, a_2_3, v0_4_2) 
plt.show()

#5
t= np.linspace(0, max_t, 100)

def x(v0,a,t):
  x = []
  for i in t:
    x.append(v0*math.cos(a)*i)
  return x

x = x(v0,a,t)

def y(v0,a,t):
  y = []
  for i in t:
    y.append(v0*math.sin(a)*i - 12*g()*i**2)
  return y

y = y(v0,a,t)
  
plt.title("5.Прыжок")
plt.xlabel("x") # ось абсцисс
plt.ylabel("y") # ось ординат
plt.grid()    
plt.plot(x,y) 
plt.show()

```

# Задание. "Число и цифра"

Запросите у пользователя целое число и сложите цифры этого числа. Повторяйте последнее действие столоко раз сколько сможете и выводите каждый раз полученную сумму.

```python
n = ''
while (True):
  n = input("Введите целое число: ")
  if n.isdigit() == False:
    print("Это не число! ")
    continue
  else:
    if int(n) != float(n) :
      print("Это не целое число! ")
    else:
      break

sum = 0
n = str(n)
for char_n in n:
  char_n = int(char_n)
  sum += char_n

print("Сумма цифр числа ", n, " = ", sum)

```
