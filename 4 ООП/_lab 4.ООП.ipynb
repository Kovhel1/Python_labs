# Упражнение №1

Создайте класс для вычисления чисел Фибоначчи. Каждое число фибоначи является объектом этого класса, которое имеет атрибуты: значение и номер. Используйте функции для инициализации (вычисления) чисел Фибоначчи как сторонние по отношению к этому классу.

```python
def Fibonacci(n):
    
    fib = [0]*max(2,n)
    fib[0] = 1
    fib[1] = 1
    for i in range(2, n):
        fib[i] = fib[i - 1] + fib[i - 2]
        
    return fib[n-1]

class Fib:
    def __init__(self, n = 1):
        self.n = n
        self.value = Fibonacci(n)
    
    def getV(self):
      return self.value
```

```python
f = Fib(8)
f.getV()
```

# Упражнение №3

Перегрузите опарации сложения, вычитания, умножения и деления для созданного класса как операции с номерами чисел Фибоначи.

```python
class Fib:

    def __init__(self, n = 1):
        self.n = n
        self.value = Fibonacci(n)

    @staticmethod
    def Fibonacci(n):
      fib = [0]*max(2,n)
      fib[0] = 1
      fib[1] = 1
      for i in range(2, n):
          fib[i] = fib[i - 1] + fib[i - 2]
          
      return fib[n-1]    
    
    def getV(self):
      return self.value

    def __add__(self, other):
      return self.getV() + other.getV()

    def __sub__(self, other):
      return self.getV() - other.getV()

    def __mul__(self, other):
      return self.getV()*other.getV()

    def __truediv__(self, other):
      if other.getV()!=0:
        return self.getV()/other.getV()
      else:
        raise Exception('delimiter is null!')
```

```python
n1 = Fib(5)
print('n1: ', n1.getV())
print('')

n2 = Fib(6)
print('n2: ', n2.getV())
print('')

print('sum: ', n1+n2)
print('sub: ', n1 - n2)
print('mul: ', n1*n2)
print('div: ', n1/n2)
```

# Задание №2. 

Составить описание класса многочленов от одной переменной, задаваемых степенью многочлена и массивом коэффициентов. Предусмотреть методы для вычисления значения многочлена для заданного аргумента, операции сложения, вычитания и умножения многочленов с получением нового объекта-многочлена, вывод на экран описания многочлена.

```python
class Polynomial():

  def __init__(self, power = 1, a = 1):
    if a == 1 or power == len(a)-1:
      self.a = a  
      self.power = power
      self.rank = len(a) #rank = power + 1, т.к нулевая степень не считается за сетепень, так удобней
    else:
      raise Exception('Количеств  множителей не соответсвует степени многочлена')
  
  def print_poly(self):
      poly = str(self.a[0])
      for i in range(1, self.rank):
        if self.a[i] != 0:
          if self.a[i] == 1:
            poly = str(poly) + '+' + 'x^' + str(i)
          else:
            poly = str(poly) + '+' + str(self.a[i]) + '*x^' + str(i)
      print(poly)

  def get_val_poly(self, x):
    pol_val = 0
    for i in range( 0, self.rank ):
      pol_val = pol_val + self.a[i]*(x**i)
    return pol_val

  def __add__(self, other):
    rank_max = max(self.rank, other.rank)
    rank_min = min(self.rank, other.rank)
    bigest_pol = self if self.rank > other.rank else other
    
    sum_a = [0]*rank_max
    for i in range(0, rank_min):
      sum_a[i] = self.a[i] + other.a[i]
    for i in range(rank_min, rank_max):
      sum_a[i] = bigest_pol.a[i]

    return Polynomial(rank_max-1, sum_a)

  def __sub__(self, other): 
    rank_max = max(self.rank, other.rank)
    rank_min = min(self.rank, other.rank)
    bigest_pol = self if self.rank > other.rank else other
    
    sub_a = [0]*rank_max
    for i in range(0, rank_min):
      sub_a[i] = self.a[i] - other.a[i]
    for i in range(rank_min, rank_max):
      sub_a[i] = bigest_pol.a[i]

    return Polynomial(rank_max - 1, sub_a)

  def __mul__(self, other): 
    
    rank_a = self.rank + other.rank
    mul_a = [0]*rank_a 
    for i in range(0, self.rank):
      for j in range(0, other.rank):
        if mul_a[i+j] != 0 :
          mul_a[i+j] = mul_a[i+j] + self.a[i]*other.a[j]
        else:
          mul_a[i+j] = self.a[i]*other.a[j]

    return Polynomial(rank_a - 1, mul_a)
  
```

```python
A = Polynomial(4, [2, 1, 0, 1, 3])
print('множители: ', A.a)
print('степень: ', A.power)
print('полином А: ')
A.print_poly()
n = 3
print('вычислим значение многочлена при заданном значении ', n, ' : ', A.get_val_poly(n))
```

```python
class Vector:

  def __init__(self, x = 0, y = 0, z = 0):
    self.x = x
    self.y = y
    self.z = z
  
  def print_vector(self):
     print(self.x, self.y, self.z)
  
  def __add__(self, other):
    x = self.x + other.x
    y = self.y + other.y
    z = self.z + other.z
    return Vector(x,y,z)

  def __sub__(self, other): 
    x = self.x - other.x
    y = self.y - other.y
    z = self.z - other.z
    return Vector(x,y,z)

  def get_len(self):
    return pow((pow(self.x,2)+pow(self.y,2)+pow(self.z,2)), 0.5)

  def get_cos(self, other):
     denominator = self.get_len()*other.get_len()
     if denominator == 0:
       return 0
     else:
       return (self.x*other.x + self.y*other.y + self.z *other.z)/denominator

  def __mul__(self, other):
      return self.get_len()*other.get_len()* self.get_cos(other)

```

```python
A = Vector(3, 5, 6)
A.print_vector()
B = Vector(7, 8, 9)
B.print_vector()
print('')

print('A lenth: ', A.get_len())
print('B lenth: ', B.get_len())
print('')

print('ABcos = ', A.get_cos(B))
print('')

C1 = A + B
C1.print_vector()
print('')

C2 = A - B
C2.print_vector()
print('')

mult = A * B
print("A*B = ", mult)
```

# Обобщённое число.

Создайте класс обобщающий понятие комплексных, двойных и дуальных чисел. И на его основе создайте классы этих чисел.

Такие числа объеденены одной формой записи:

$$ c = a + ib $$

где c - обобщённое число (комплексное, двойное или дуальное), a и b - вещественные числа, i - некоммутирующий символ.

Именно из-за наличия символа i число c не просто сумма a и b. Такие числа можно представлять как вектор на плоскости (a,b).

А символ i обладает следующим свойством:

- для комплексных чисел

$$ i^2 = -1 $$

- для двойных чисел

$$ i^2 = 1 $$

- для дуальных чисел

$$ i^2 = 0 $$

Перегрузить для них базовые операции: сложения, вычитания, умножения и деления.

Например, операция умножения для таких чисел имеет вид:

$$ (a_1+b_1i)\cdot (a_2+b_2i)=a_1a_2+b_1a_2i+a_1b_2i+b_1b_2i^{2}=(a_1a_2+b_1b_2i^{2})+(b_1a_2+a_1b_2)i. $$

```python
class Numbers:
  def __init__(self, a = 0,b = 0):
    self.a = a
    self.b = b

  def __add__(self, other):
    return Numbers(self.a + other.a, self.b + other.b)

  def __sub__(self, other):
    return Numbers(self.a - other.a, self.b - other.b)

  def __mul__(self, other):
    mult = [self.a*other.a, self.b*other.a + self.a*other.b, self.b*other.b]
    return mult
```

```python
C = Numbers(5,6)
D = Numbers(7,8)
L = C + D
print("L.a = ", L.a)
print("L.b = ", L.b)
print('')

L = C - D
print("L.a = ", L.a)
print("L.b = ", L.b)
print('')

print("C*D = ", C*D)
```

```python
class Complex(Numbers):

  i2  = -1

  def __mul__(self, other):
    return Complex(self.a*other.a+self.b*other.b*self.i2, self.b*other.a+self.a*other.b)

  def __truediv__(self, other):
    denum = other.a*other.a + other.b*other.b
    return Complex( (self.a*other.a + self.b*other.b)/denum,
                    (self.b*other.a - self.a*other.b)/denum )

class Double(Numbers):

  i2 = 1

  def __mul__(self, other):
    return Double(self.a*other.a+self.b*other.b*self.i2, self.b*other.a+self.a*other.b)

  def __truediv__(self, other):
    denum = self.a*self.a - self.b*self.a
    return Double( (self.a*other.a - self.b*other.b)/denum,
                  (-other.a*self.b - self.a*other.b)/denum  )

class Dual(Numbers):

  i2 = 0

  def __mul__(self, other):
    return Dual(self.a*other.a+self.b*other.b*self.i2, self.b*other.a+self.a*other.b)

  def __truediv__(self, other):
    return Dual( self.a/other.b, (other.a*self.b - self.a*other.b)/other.a*other.a )
```

```python
a = 12
b = 9

A1 = Complex(a,b)
B1 = Complex(b,a)
C1 = A1 + B1
print("Complex +: ", C1.a, C1.b)
C1 = A1 - B1
print("Complex -: ", C1.a, C1.b)
C1 = A1 * B1
print("Complex *: ", C1.a, C1.b)
C1 = A1 / B1
print("Complex /: ", C1.a, C1.b)
print('')

A2 = Double(a,b)
B2 = Double(b,a)
C2 = A2 + B2
print("Double +: ", C2.a, C2.b)
C2 = A2 - B2
print("Double -: ", C2.a, C2.b)
C2 = A2 * B2
print("Double *: ", C2.a, C2.b)
C2 = A2 / B2
print("Double /: ", C2.a, C2.b)
print('')

A3 = Dual(a,b)
B3 = Dual(b,a)
C3 = A3 + B3
print("Dual +: ", C3.a, C3.b)
C3 = A3 - B3
print("Dual -: ", C3.a, C3.b)
C3 = A3 * B3
print("Dual *: ", C3.a, C3.b)
C3 = A3 / B3
print("Dual /: ", C3.a, C3.b)
print('')
```
